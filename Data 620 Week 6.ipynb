# Data 620 Week 6 - Southern Women Data

# Intro
We are given a dataset that shows a simple 2-node network looking at the attendance of 18 Southern Women at 14 social events. Using Network X we're looking to see relationships between (1) the women, and (2) the social events. 

```python
import networkx as nx
import numpy as np
import pandas as pd
```

# Load Data 

```python
G = nx.davis_southern_women_graph()

#Separate the two sections
women = G.graph['top']
clubs = G.graph['bottom']
```

# Create Network Graph

```python
nx.draw(G, node_size=400,
        with_labels=True, alpha=.75, font_size=15)
```

# Bipartite Projection: Women
There will be a connection between the women if they participate in the same club/social event. We can see connections between women using Network X.

```python
W = nx.bipartite.weighted_projected_graph(G, women)

nx.draw(W, node_size=800,
        with_labels=True, node_color='b', alpha=.5, font_size=15)
```

# Degree Centrality 

```python
nx.degree_centrality(W)
```

#  Betweeness Centrality

```python
nx.betweenness_centrality(W)
```

#  Closeness Centrality

```python
nx.closeness_centrality(W)
```

# Bipartite Projection: Events

```python
C = nx.bipartite.weighted_projected_graph(G, clubs)
nx.draw(C, node_size=800,
        with_labels=True, node_color='g', alpha=.5, font_size=15)
```

# Events Degrees

```python
nx.degree(C)
```
